```python
#data cleansing
dataindo=data.loc[data['location']=='Indonesia']
dataindo=dataindo.reset_index(drop=True)
```

```python
dataindo=dataindo[['date','total_cases','new_cases','total_deaths','new_deaths']]
```

```python
#ganti data nan menjadi 0
dataindo=dataindo.fillna(0)
```

```python
#EDA
dataindo['new_cases'].plot(label='Kasus baru')
dataindo['new_deaths'].plot(label='kasus kematian')
plt.xlabel('hari ke=')
plt.ylabel('kasus perhari')
plt.legend(loc='best')
plt.show()
```
